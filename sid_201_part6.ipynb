# plot cumulative

```python
csum = infection_trace_df.\
    groupby(['rep', 'infection_time', 'tr'])['transmitee_pid'].\
    count().\
    reset_index()
```

```python
csum['cumulative'] = csum.groupby(['rep', 'tr'])['transmitee_pid'].cumsum()
```

```python
csum
```

```python
# plot cumulative

sns.lmplot(x = 'infection_time', y = 'cumulative',
           col = 'rep', row = 'tr',
           fit_reg = False, data = csum)
```

```python
# plot average rep

df = csum.groupby(['rep', 'tr', 'infection_time'])['cumulative'].mean().reset_index()

sns.lmplot(x = 'infection_time', y = 'cumulative',
           hue = 'rep', col = 'tr',
           fit_reg = False, data = df)
```

```python
# plot average rep

df = csum.groupby(['tr', 'infection_time'])['cumulative'].mean().reset_index()

sns.lmplot(x = 'infection_time', y = 'cumulative',
           hue='tr', fit_reg = False, data = df)
```

# extra stuff

```python
# make prettier
sns.tsplot(time = 'infection_time', value = 'cumulative',
           condition = 'tr', unit = 'tr', data = df)
```
