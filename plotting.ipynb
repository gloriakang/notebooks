```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
```

```python
df = pd.read_csv("aa_0_ai.csv")
```

```python
df
```

```python
# count number of infections by day
df1 = df.groupby('intervened')['ig_type'].count()
print(df1)
```

```python
# remove -1's
df1 = df[df['intervened'] != -1]
df1 = df1.groupby('intervened')['ig_type'].count()
```

```python
# plot infections by day
plt.plot(df1)
plt.xlabel('day intervened')
plt.ylabel('count')
plt.title('infections per day')
```

```python
# cumulative
df1 = df[df['intervened'] != -1]
df1 = df1.groupby('intervened')['ig_type'].count().cumsum()
```

```python
# plot cumulative infections
plt.plot(df1)
plt.xlabel('day intervened')
plt.ylabel('count')
plt.title('cumulative infections')
```

```python
# histogram
df1 = df[df['intervened'] != -1]
df1 = df1.groupby('intervened')['ig_type'].count()
df1.plot.hist()
plt.xlabel('day intervened')
plt.title('histogram')
```
